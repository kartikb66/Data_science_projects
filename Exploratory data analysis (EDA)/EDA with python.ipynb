```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
student_df = pd.read_csv('StudentsPerformance.csv')
student_df.head()
```

##  Cleaning data

```python
# checking for null values
student_df.isnull().sum()
```

```python
# checking for NA avlues
student_df.isna().sum()
```

```python
student_df = student_df.drop('race/ethnicity', axis = 1)
student_df.head()
```

```python
student_df = student_df.drop('parental level of education', axis = 1)
student_df.head()
```

## Relationship Analysis 

```python
sns.pairplot(student_df)
```

```python
sns.scatterplot(student_df)
```

```python
sns.relplot(x = 'math score', y = 'reading score', hue = 'gender', data = student_df)
```

```python
sns.distplot(student_df['math score'])
```
